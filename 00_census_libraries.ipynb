Class for representing Census geographies
- class censusdata.censusgeo.censusgeo(geo, name='')


Parameters:
- geo (tuple of 2-tuples of strings) – Tuple of 2-tuples of the form (geographic component, identifier), where geographic component is a string (e.g., ‘state’) and identifier is either a numeric code (e.g., ‘01’) or a wildcard (‘*’). These identify the geography in question.
- name (str, optional) – Name of geography (e.g., ‘Alabama’).

Functions for downloading data and lists of geographies from the Census API.

censusdata.download.download(src, year, geo, var, key=None, tabletype='detail', endpt='')

Parameters:	
1. src (str) – Census data source: ‘acs1’ for ACS 1-year estimates, ‘acs5’ for ACS 5-year estimates, ‘acs3’ for ACS 3-year estimates, ‘acsse’ for ACS 1-year supplemental estimates, ‘sf1’ for SF1 data.
2. year (int) – Year of data.
geo (censusgeo) – Geographies for which to download data.
3. var (list of str) – Census variables to download.
4. key (str, optional) – Census API key.
5. tabletype (str, optional) – Type of table from which variables are drawn (only applicable to ACS data). Options are ‘detail’ (detail tables), ‘subject’ (subject tables), ‘profile’ (data profile tables), ‘cprofile’ (comparison profile tables).
6. endpt (str, optional) – Allows override of whether old or new API endpoint is used. Specify ‘old’ for old, ‘new’ for new, ‘’ to use default. This option generally shouldn’t need to be specified but can be helpful if download problems are encountered.
- Returns: pandas.DataFrame ... Data frame with columns corresponding to designated variables, and row index of censusgeo objects representing Census geographies.


```python
# Download ACS 2011-2015 5-year estimates for Oakland city, California on population size, median age, and median household income.
cd.download('acs5', 2015, cd.censusgeo([('state', '06'), ('place', '53000')]), ['B01001_001E', 'B01002_001E', 'B19013_001E'])
```
